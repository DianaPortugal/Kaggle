```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df=pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
df
```

```python
df.info()
```

```python
dfs=pd.read_csv("/kaggle/input/digit-recognizer/sample_submission.csv")
dfs
```

```python
x = df.drop("label", axis=1)
y = df["label"]
```

```python
dft=pd.read_csv("/kaggle/input/digit-recognizer/test.csv")
```

```python
from sklearn.neighbors import KNeighborsClassifier

modelo = KNeighborsClassifier(n_neighbors=3)
modelo.fit(x, y)

predictions = modelo.predict(dft)
predictions
```

```python
submision=pd.DataFrame({"ImageId":dfs["ImageId"],"Label":predictions})
submision
```

```python
submision.to_csv("submission.csv", index=False)
```
