```python
import pandas as pd
```

```python
df = pd.read_csv("/train.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df2 = pd.read_csv("/test.csv")
df2.info()
```

```python
dfa = df.drop("PassengerId", axis = 1)
dfa.info()
```

```python
dfb = dfa.drop("Cabin", axis = 1)
dfb.info()
```

```python
dfc = dfb.drop("Name", axis = 1)
dfc.info()
```

```python
dfc["Ticket"].nunique()
```

```python
dfd = dfc.drop("Ticket", axis = 1)
dfd.info()
```

```python
dfd.head()
```

```python
dfd["Embarked"].value_counts()
```

```python
dfd["Embarked"].fillna(dfd["Embarked"].mode()[0], inplace = True)
dfd.info()
```

```python
dfd["Age"].fillna(dfd["Age"].mean(), inplace=True)
dfd.info()
```

```python
dfd.head()
```

```python
dfe = pd.get_dummies(dfd, columns=["Sex"], prefix="Sex")
dfe.head()
```

```python
dff = pd.get_dummies(dfe, columns=["Embarked"], prefix="Embarked")
dff.head()
```

```python
y = dff["Survived"]
x = dff.drop("Survived", axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x,y)
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2a = df2.drop("PassengerId", axis = 1)
df2a.info()
```

```python
df2b = df2a.drop("Name", axis = 1)
df2b.info()
```

```python
df2c = df2b.drop("Ticket", axis = 1)
df2c.info()
```

```python
df2d = df2c.drop("Cabin", axis = 1)
df2d.info()
```

```python
df2d.head()
```

```python
df2d["Age"].fillna(df2d["Age"].mean(), inplace=True)
df2d.info()
```

```python
df2d["Fare"].fillna(df2d["Fare"].mean(), inplace=True)
df2d.info()
```

```python
df2f = pd.get_dummies(df2d, columns=["Sex"], prefix="Sex")
df2f.head()
```

```python
df2e = pd.get_dummies(df2f, columns=["Embarked"], prefix="Embarked")
df2e.head()
```

```python
solutions = lr.predict(df2e)
solutions
```

```python
gs=pd.read_csv("/gender_submission.csv")
gs.head()
```

```python
submission = pd.DataFrame({'PassengerId':df2['PassengerId'], 'Survived':solutions})
submission.to_csv('submission.csv',index=False)
```
